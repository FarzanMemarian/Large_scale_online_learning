```python
import numpy as np 
import time
# Generating data 

Nsamp = 10**5
Ndim =20
X_c1 = np.random.multivariate_normal(mean= np.ones(Ndim), cov = np.identity(Ndim),size = Nsamp)
X_c2 = np.random.multivariate_normal(mean= -np.ones(Ndim), cov = np.identity(Ndim),size = Nsamp)
Y_c1 = np.ones(Nsamp)
Y_c2= -np.ones(Nsamp)

X = np.concatenate((X_c1,X_c2))
Y = np.concatenate((Y_c1,Y_c2))

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=10**5)

# Generating permutations of the training set 
Nper = 30
X_tr_perm=[]
Y_tr_perm=[]

for i in range(Nper):
    ind = np.random.permutation(Nsamp)
    X_tr_perm.append(X_train[ind])
    Y_tr_perm.append(y_train[ind])
                             
```

```python
def online_kalman(x,y,Nsamp):
    # Online Kalman Filter 

    phi = np.identity(Ndim) # Initializing as identity 
    theta = np.random.uniform(-0.5,0.5,size=Ndim) # Parameter vector has dimension Ndim 

    start_time = time.time()

    for j in range(Nsamp):
        tau = max(20,j-40)
        f = 1.71*np.tanh(0.66*x.dot(theta))
        df =  1.71*0.66*(1 - np.tanh(0.66*x.dot(theta))**2)
        phi=np.linalg.inv((1-2/tau)*np.linalg.inv(phi)+ (2/tau)*(df[j]*df[j])*(np.outer(x[j,:],x[j,:])))

        dL=x[j,:]*(-2*df[j]*(1.5*y[j]-f[j]))

        theta = theta - (1/tau)*phi.dot(dL)
        
    time_o_run= (time.time() - start_time)

    print("--- %s seconds ---" % (time_o_run))
    
    return (theta,time_o_run)
```

```python
#Finding optimal theta using Batch Newton on test data

x=X_test
y=y_test

# Batch Newton Method

theta = np.random.uniform(-0.5,0.5,size=Ndim) # Parameter vector has dimension Ndim 


start_time = time.time()

# Although high memory, absolutely necessary to prevent floating point errors resulting in singular H 
dL=np.empty(shape=(Nsamp,Ndim))
g=np.empty(shape=(1,Ndim))
H=np.empty(shape=(Ndim,Ndim))
h=np.empty(shape=(Ndim,Ndim,Nsamp))

N_iter=0
while True:
    theta_old=theta
    
    f = 1.71*np.tanh(0.66*x.dot(theta))
    df =  1.71*0.66*(1 - np.tanh(0.66*x.dot(theta))**2)

    for j in range(Nsamp):
        dL[j,:] = x[j,:]*(-2*df[j]*(1.5*y[j]-f[j]))
        h[:,:,j]=(df[j]*df[j])*(np.outer(x[j,:],x[j,:]))

    g=np.sum(dL,axis=0)
    H=np.sum(h,axis=2)
    theta=theta-np.linalg.inv(H).dot(g)
    
    N_iter=N_iter+1
    print(np.linalg.norm(theta-theta_old))
    #0.01/Nsamp should be the tolerance. Does not converge if initial diff is > 3. Use 0.4 for 10^5
    if np.linalg.norm(theta-theta_old)< (0.4):
        break
        
print("--- %s seconds ---" % (time.time() - start_time))
print(N_iter)

theta_opt=theta
print(theta_opt)
```

```python
# Training using different number of examples from the permuted sets for Online Kalman algorithm 

Nsamp_perm = [x for x in range(3000,10**5,3000)]
Nsamp_perm[-1]=10**5

theta_list=[]
time_list=[]
for x,y,Nsamp in zip(X_tr_perm,Y_tr_perm,Nsamp_perm):
    x=x[:Nsamp,:]
    y=y[:Nsamp]
    theta_train,train_time = online_kalman(x,y,Nsamp)

    theta_list.append(theta_train)
    time_list.append(train_time)

print(time_list)
print(theta_list)
```

```python
err = np.empty(len(theta_list))
for n in range(len(theta_list)):
    err[n]=np.linalg.norm(theta_list[n]-theta_opt)**2
    
import matplotlib.pyplot as plt
plt.plot(Nsamp_perm[:len(theta_list)], err, 'ko-')
plt.ylabel(r'($\theta - \theta*)^2$')
plt.xlabel('Number of examples')
plt.show()

plt.plot(time_list[:len(theta_list)], err, 'ko-')
plt.ylabel(r'($\theta - \theta*)^2$')
plt.xlabel('Time of execution')
plt.show()
```
